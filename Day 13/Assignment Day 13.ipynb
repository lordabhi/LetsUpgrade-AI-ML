## Step 5 – Statistical Tests (Mann-Whitney)


### Attrition Vs Distance from Home 

```python
from sklearn import preprocessing
Label_encoder = preprocessing.LabelBinarizer()
dataset.Attrition = Label_encoder.fit_transform(dataset.Attrition)
dataset.Attrition
```

```python
import pandas as pd
datasetYes = pd.read_excel("Attrition_Yes_No.xls", sheet_name="Yes")
datasetYes.head()
```

```python
datasetNo = pd.read_excel("Attrition_Yes_No.xls", sheet_name= "No")
datasetNo.head()
```

```python
from scipy.stats import mannwhitneyu  
```

```python
a1 = datasetYes.DistanceFromHome
a2 = datasetNo.DistanceFromHome
stat,p = mannwhitneyu(a1,a2)
print(stat,p)
```

#### H0: There is no significant differences in the Distance From Home between attrition (Y) and attirition (N) 
#### Ha: There is significant differences in the Distance From Home between attrition (Y) and attirition (N) 
#### As the P value of 0.46 is > 0.05, the Ha is rejected and H0 is accepted.

## Statistical Tests (Separate T Test) 

```python
from scipy.stats import ttest_ind
```

### Attrition Vs Distance From Home

```python
z1 = datasetYes.DistanceFromHome
z2 = datasetNo.DistanceFromHome
stat,p = ttest_ind(z2,z1)
print(stat,p)
```

#### As the P value is 0.518286042805572, which is > than 0.05, the HA is rejected and H0 is accepted.  
#### H0: There is no significant differences in the Distance From Home between attrition (Y) and attirition (N) 
####  HA: There is significant differences in the Distance From Home between attrition (Y) and attirition (N)

## Unsupervised Learning - Correlation Analysis 

```python
from scipy.stats import pearsonr 
```

```python
dataset['TotalWorkingYears']=dataset['TotalWorkingYears'].fillna(11.28) 
```

```python
stats, p=pearsonr(dataset.Attrition, dataset.DistanceFromHome) 
print(stats, p) 
```

```python
stats, p=pearsonr(dataset.Attrition, dataset.MonthlyIncome) 
print(stats, p) 
```

```python
stats, p=pearsonr(dataset.Attrition, dataset.TotalWorkingYears) 
print(stats, p) 
```

```python
stats, p=pearsonr(dataset.Attrition, dataset.YearsAtCompany) 
print(stats, p) 
```

```python
stats, p=pearsonr(dataset.Attrition, dataset.YearsWithCurrManager) 
print(stats, p) 
```

### The inference of the above analysis are as follows: 

### Attrition & DistanceFromHome:  
#### As r = -0.009, there’s low negative correlation between Attrition and DistanceFromHome 
#### As the P value of 0.518 is > 0.05, we are accepting H0 and hence there’s no significant correlation 
#### between Attrition & DistanceFromHome
